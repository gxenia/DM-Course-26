## Pandas Data Structures

In order to get started with Pandas, you need to get comfortable with its two workhorse
data structures: Series and DataFrame.

### Series
A Series is a one-dimensional array-like object containing an array of data (of any Numpy data type) and an associated array of data labels, called its index.

```python
obj = Series(np.arange(10))
describe(obj)
```

Please note the main difference with numpy arrays. A Series object is associated with an object of type index (printed  on the left side).


Since in the previous case we did not specify any index, by default pandas assign all the integers within the interval $[0, N)$, where $N$ is the length of the series.

We can also access to index ad values of a Series object, independently from each other, as the following:

```python
print(obj.index, obj.values)
```

It is often useful to create a Series with an index associated with each data point.

You can specify an index while instantiating the Series:

```python
obj = Series([1,2,3,4], index=list('abcd'))
describe(obj)
print(obj['b'])
```

It should be noted that a Series is very similar to a fixed-length, ordered dict.
In fact a Series is also a mapping of index to data values. 

It can be therefore replaced into many functions that expect a
dict:

```python
obj = Series(
    [1,2,3,4,5,6,7],
    index=np.array(['Bob', 'Joe', 'Will', 'Bob', 'Will', 'Joe', 'Joe']),
)

# Note: 1 key multiple values
print(obj['Joe']) # access like a dict 
print(hrule(20))

print(obj.Joe)  # access as an object field
print(hrule(20))

# test membership
print('Joe' in obj)
print('Frank' in obj)
```

It is also straightforward to construct a Series from a Python dict. 

```python
sdata = {'Ohio': 35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000}
obj3 = Series(sdata)
print(obj3)
```

The index is ordered by default. However we can also specify the exact order we want.

```python
states = ['California', 'Ohio', 'Oregon', 'Texas']

obj = Series(sdata, index=states) 
describe(obj)
```

### DataFrame = "Cornice per i dati"

E' una concatenazione di series, le quali a loro volta sono vettori numpy + indice.

A DataFrame represents a tabular, spreadsheet-like data structure an ordered collection of columns, each of which can be a different value type.

A DataFrame has both a row and column index.

```python
# each key will correspond to a column of the DataFrame
data = {
    'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada'],
    'year': [2000, 2001, 2002, 2001, 2002],
    'pop': [1.5, 1.7, 3.6, 2.4, 2.9]
}
df = DataFrame(data)  # Note: the index is automatically assigned 
describe(df)
```

Likewise the series, we can rearrange the columns of the dataframe

```python
df = DataFrame(data, columns=['year', 'state', 'pop'])
describe(df)
```

To access the column of a dataframe we can use a dict-like notation or an attribute-like one.

```python
describe(df.state)
print(hrule(20))
describe(df['state'])
```

Column indexing can also be used for adding a new column

```python
df['density'] = 10 # broadcast -> columns are numpy arrays
describe(df)
```

**Note**: Remember that the Series object returned by ``df[columnName]`` is always *view* on the original data, thus any change to that object will be reflected
on the dataframe (unless we call copy() the Series)


NOTA 2: adesso non si può più accedere alle righe con l'accesso diretto quindi la modifica della vista non è consentita

```python
df['state'][0] = "New York"
```

Per accedere alle righe esistono due metodi `loc` e `iloc`.
- `loc` si basa sull'indice
- `iloc` si basa sulla posizione della riga (infatti `iloc` sta per *index location*)

```python
df = DataFrame(data, columns=['year', 'state', 'pop'], index=np.flip(np.arange(0, len(next(iter(data)))))) # LOOK: the index is in reverse order
df
```

# Data Loading, Storage, and File Formats
Pandas features a number of functions for reading tabular data as DataFrame object like `read_table`, `read_csv`, `read_json` and `read_html`. Some frequently used options related to these methods are:
- `path` - string indicating file system location, URL or file-like object
- `sep`, `delimiter` - character sequence or regex to use to split fields in each ros
- `header` - row number to use as column names; defaults to 0 (first row) but should be None if there is no header rows. If we pass None we can specify the names of the columns with the `names` parameter (if we don't pass them pandas assigns numbers 0 to num_col-1)
- `skiprows` - number of rows at beginning of file to ignore or list of row numbers to skip

## Data Cleaning and Preparation

#### Filtering Missing Data

While filtering missing data we can either use the mask returned by the ``isnull``
(there is also a method ``notnull``) or use the ``dropna`` function.
Here is an example:

```python
not_null = obj.dropna() # it returns a copy
describe(not_null) # the row at index 2 has been removed
```

**Example**:  an equivalent method for removing null values from obj

Hint: Boolean indexing

```python
describe(
    obj[~obj.isnull()]
)
```

``dropna`` works also with dataframes. The only difference with respect to Series is that you
need also to provide the axis along which you want to perform the operation.

You can either remove rows or columns (entirely)

By default pandas remove an entire record (a row or column) as it finds even just a single missing value.

As a consequence this single missing value can cause a massive loss of information. 
You override this behavior by specifying a different value for the ``how`` argument of dropna (by default it is ``'any'``). With *how='all'* you are telling to pandas that an entry can be regarded as NaN only if all the attributes have a NaN value.

Rather than removing entire records we can decide to fill the missing values.

The instruction is ``fillna``
